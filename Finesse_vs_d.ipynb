```python
import numpy as np
import matplotlib.pyplot as plt

# All efficiency models are normalized between 1 and 0, to get percentage multiply by 100

import numpy as np

## AFC Efficiency for backwards propagation Gaussian peaks
def AFC_gb(d_tilde,F,d0):
    eta = (1 - np.exp(-d_tilde))**2 * np.exp(-7/F**2) * np.exp(-d0)
    return eta

## AFC Efficiency for forwards propagation Gaussian peaks
def AFC_gf(d_tilde,F,d0):
    eta = d_tilde**2 * np.exp(-d_tilde) * np.exp(-7/F**2) * np.exp(-d0)
    return eta

## AFC Efficiency for backwards propagation square peaks
def AFC_sb(d_tilde,F,d0):
    eta = (1 - np.exp(-d_tilde))**2 * (np.sinc(1/F))**2 * np.exp(-d0)
    return eta

## AFC Efficiency for forwards propagation square peaks
def AFC_sf(d_tilde,F,d0):
    eta = d_tilde**2 * np.exp(-d_tilde) * (np.sinc(1/F))**2 * np.exp(-d0)
    return eta
```

```python
d0 = 0
F = np.linspace(1,10,50)
d = np.linspace(1,10,50)

d_tilde = np.zeros((len(d),len(F)))

Fv, dv = np.meshgrid(F,d)

## Plotting Figures ##--------------------------------------------

# 3d surface plots 
# Gaussian Peaks
fig1 = plt.figure()
ax1 = fig1.add_subplot(121, projection='3d')
ax2 = fig1.add_subplot(122, projection='3d')
fig1.suptitle(r'Gaussian AFC Peaks @ $d_{0} = %.1f$' %d0)
fig1.subplots_adjust(top=0.915, bottom=0.1, left=0.08, right=0.98, hspace=0.2, wspace=0.2)

ax1.plot_surface(Fv, dv, 100*AFC_gf(dv/Fv,F,d0), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax1.view_init(15,-135)
ax1.set_title('Forward Propagation')
ax1.set_xlabel('Finesse F')
ax1.set_ylabel('Peak Absorbtion Depth d')
ax1.set_zlabel('Efficiency %')

ax2.plot_surface(Fv, dv, 100*AFC_gb(dv/Fv,F,d0), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax2.view_init(15,-135)
ax2.set_title('Backwards Propagation')
ax2.set_xlabel('Finesse F')
ax2.set_ylabel('Peak Absorbtion Depth d')
ax2.set_zlabel('Efficiency %')
```

```python
fig2 = plt.figure()
ax1 = fig2.add_subplot(121)
ax2 = fig2.add_subplot(122)
fig2.suptitle(r'Gaussian AFC Peaks @ $d_{0} = %.1f$' %d0)
fig2.subplots_adjust(top=0.85, bottom=0.11, left=0.125, right=0.9, hspace=0.2, wspace=0.6)

ax1.contourf(Fv, dv, 100*AFC_gf(dv/Fv,F,d0))
ax1.set_title('Forwards Propagation')
ax1.set_xlabel('Finesse F')
ax1.set_ylabel('Peak Absorbtion Depth d')
plt.colorbar(ax1.contourf(Fv, dv, 100*AFC_gf(dv/Fv,F,d0)), ax = ax1)

ax2.contourf(Fv, dv, 100*AFC_gb(dv/Fv,F,d0))
ax2.set_title('Backwards Propagation')
ax2.set_xlabel('Finesse F')
ax2.set_ylabel('Peak Absorbtion Depth d')
plt.colorbar(ax2.contourf(Fv, dv, 100*AFC_gb(dv/Fv,F,d0)), ax = ax2)
```

```python
fig3 = plt.figure()
ax1 = fig3.add_subplot(121, projection='3d')
ax2 = fig3.add_subplot(122, projection='3d')
fig3.suptitle(r'Square AFC Peaks @ $d_{0} = %.1f$' %d0)
fig3.subplots_adjust(top=0.915, bottom=0.1, left=0.08, right=0.98, hspace=0.2, wspace=0.2)

ax1.plot_surface(Fv, dv, 100*AFC_sf(dv/Fv,F,d0), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax1.view_init(15,-135)
ax1.set_title('Forward Propagation')
ax1.set_xlabel('Finesse F')
ax1.set_ylabel('Peak Absorbtion Depth d')
ax1.set_zlabel('Efficiency %')

ax2.plot_surface(Fv, dv, 100*AFC_sb(dv/Fv,F,d0), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax2.view_init(15,-135)
ax2.set_title('Backward Propagation')
ax2.set_xlabel('Finesse F')
ax2.set_ylabel('Peak Absorbtion Depth d')
ax2.set_zlabel('Efficiency %')
```
