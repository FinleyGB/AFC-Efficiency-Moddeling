```python
import numpy as np
import matplotlib.pyplot as plt

# All efficiency models are normalized between 1 and 0, to get percentage multiply by 100

import numpy as np

## AFC Efficiency for backwards propagation Gaussian peaks
def AFC_gb(d_tilde,F,d0):
    eta = (1 - np.exp(-d_tilde))**2 * np.exp(-7/F**2) * np.exp(-d0)
    return eta

## AFC Efficiency for forwards propagation Gaussian peaks
def AFC_gf(d_tilde,F,d0):
    eta = d_tilde**2 * np.exp(-d_tilde) * np.exp(-7/F**2) * np.exp(-d0)
    return eta

## AFC Efficiency for backwards propagation square peaks
def AFC_sb(d_tilde,F,d0):
    eta = (1 - np.exp(-d_tilde))**2 * (np.sinc(1/F))**2 * np.exp(-d0)
    return eta

## AFC Efficiency for forwards propagation square peaks
def AFC_sf(d_tilde,F,d0):
    eta = d_tilde**2 * np.exp(-d_tilde) * (np.sinc(1/F))**2 * np.exp(-d0)
    return eta
```

```python
## Efficiency vs Finesse ##

F = np.linspace(1,10,100)        # Finesse
d = np.array([1,3,5,7,10])    # Peak absorbions depth
d0 = 0.5                         # Background absorbtion
d_tilde_5 = d[0]/F               # Effective absorbtion @ d = 5
d_tilde_10 = d[1]/F              # Effective absorbtion @ d = 10
d_tilde_15 = d[2]/F              # Effective absorbtion @ d = 15
d_tilde_20 = d[3]/F              # Effective absorbtion @ d = 20
d_tilde_30 = d[4]/F              # Effective absorbtion @ d = 30

## Gaussian AFC Peaks ##

fig1 ,(ax1, ax2) = plt.subplots(1, 2)
fig1.suptitle('Gaussian AFC Peaks @ $d_{0} = %.1f$' %d0)

ax1.plot(F, 100*AFC_gf(d_tilde_5,F,d0), label = 'd = %d' %d[0])
ax1.plot(F, 100*AFC_gf(d_tilde_10,F,d0), label = 'd = %d' %d[1])
ax1.plot(F, 100*AFC_gf(d_tilde_15,F,d0), label = 'd = %d' %d[2])
ax1.plot(F, 100*AFC_gf(d_tilde_20,F,d0), label = 'd = %d' %d[3])
ax1.plot(F, 100*AFC_gf(d_tilde_30,F,d0), label = 'd = %d' %d[4])

ax1.set_title('Forwards Propagation')
ax1.set_ylabel('Efficiency (%)')
ax1.set_xlabel('Finesse F')
ax1.set_ylim(0,100)
ax1.grid('true')
ax1.legend(loc = 'upper left')


ax2.plot(F,100*AFC_gb(d_tilde_5,F,d0))
ax2.plot(F,100*AFC_gb(d_tilde_10,F,d0))
ax2.plot(F,100*AFC_gb(d_tilde_15,F,d0))
ax2.plot(F,100*AFC_gb(d_tilde_20,F,d0))
ax2.plot(F,100*AFC_gb(d_tilde_30,F,d0))

ax2.set_title('Backwards Propagation')
ax2.set_ylabel('Efficiency (%)')
ax2.set_xlabel('Finesse F')
ax2.set_ylim(0,100)
ax2.grid('true')

fig1.subplots_adjust(top = 0.85, wspace = 0.4)
```

```python
## Square AFC Peaks ##

fig2 ,(ax1, ax2) = plt.subplots(1, 2)
fig2.suptitle('Square AFC Peaks @ $d_{0} = %.1f$' %d0)

ax1.plot(F, 100*AFC_sf(d_tilde_5,F,d0), label = 'd = %d' %d[0])
ax1.plot(F, 100*AFC_sf(d_tilde_10,F,d0), label = 'd = %d' %d[1])
ax1.plot(F, 100*AFC_sf(d_tilde_15,F,d0), label = 'd = %d' %d[2])
ax1.plot(F, 100*AFC_sf(d_tilde_20,F,d0), label = 'd = %d' %d[3])
ax1.plot(F, 100*AFC_sf(d_tilde_30,F,d0), label = 'd = %d' %d[4])
ax1.legend(loc = 'upper left')

ax1.set_title('Forwards Propagation')
ax1.set_ylabel('Efficiency (%)')
ax1.set_xlabel('Finesse F')
ax1.set_ylim(0,100)
ax1.grid('true')


ax2.plot(F,100*AFC_sb(d_tilde_5,F,d0))
ax2.plot(F,100*AFC_sb(d_tilde_10,F,d0))
ax2.plot(F,100*AFC_sb(d_tilde_15,F,d0))
ax2.plot(F,100*AFC_sb(d_tilde_20,F,d0))
ax2.plot(F,100*AFC_sb(d_tilde_30,F,d0))

ax2.set_title('Backwards Propagation')
ax2.set_ylabel('Efficiency (%)')
ax2.set_xlabel('Finesse F')
ax2.set_ylim(0,100)
ax2.grid('true')

fig2.subplots_adjust(top = 0.85, wspace = 0.4)
```
