Modelling the efficiency equations for the AFC storage from papers: 

1. Afzelius M, Simon C, De Riedmatten H, Gisin N. Multimode quantum memory based on atomic frequency combs. Physical Review A. 2009 May 21;79(5):052329.
4. Jobez P, Usmani I, Timoney N, Laplane C, Gisin N, Afzelius M. Cavity-enhanced storage in an optical spin-wave memory. New Journal of Physics. 2014 Aug 4;16(8):083005.

```python
import matplotlib.pyplot as plt
import numpy as np

def AFC_nu_0(d_tilda,d0,F):     # Efficiency from whitebaord meeting
    nu = d_tilda**2 * np.exp(-7/F) * np.exp(-d_tilda) * np.exp(-d0)
    return nu

def AFC_nu_1(d_tilda,F):   #Efficiency from paper 1
    nu = (1 - np.exp(-d_tilda))**2 * np.exp(-7/(F**2))
    return nu

def AFC_nu_4(d_tilda,d0,F):     #Efficiency from paper 4
    nu = d_tilda**2 * np.exp(-d_tilda) * (np.sinc(np.pi/F))**2 * np.exp(-d0) # d0 term is not included in paper
    return nu
```

```python
d_tilda = 2 # absorbtion coefficient
L = 0.5       # medium length
d0 = 0      # background absorbtion
F = np.linspace(1,50,100) # finesse

fig1, (ax1, ax2, ax3) = plt.subplots(1,3)
#fig1.suptitle(r'Efficiency $\eta_{AFC}$ vs Finesse F')

ax1.plot(F,AFC_nu_0(d_tilda,d0,F)*100)
ax1.set_title(r'$\eta_{AFC} = \tilde{d}^2 e^{\tilde{-d}} e^{\frac{-7}{F}} e^{{-d_0}}$')
ax1.set_xlabel('Finesse')
ax1.set_ylabel('Efficiency (%)')
ax1.grid('true')

ax2.plot(F,AFC_nu_1(d_tilda,F)*100)
ax2.set_title(r'$\eta_{AFC} = (1 - e^{\tilde{-d}})^2 e^{\frac{-7}{F^2}}$')
ax2.set_xlabel('Finesse')
ax2.set_ylabel('Efficiency (%)')
ax2.grid('true')

ax3.plot(F,AFC_nu_4(d_tilda,d0,F)*100)
ax3.set_title(r'$\eta_{AFC} = \tilde{d}^2 e^{\tilde{-d}} sinc(\frac{\pi}{F})^2 e^{{-d_0}}$')
ax3.set_xlabel('Finesse')
ax3.set_ylabel('Efficiency (%)')
ax3.grid('true')
```

```python
d_tilda = np.linspace(1,50,100)
L = 1
d0 = 0
F = 5

fig2, (ax1, ax2, ax3) = plt.subplots(1,3)
#fig2.suptitle(r'Efficiency $\eta_{AFC}$ vs Absorbtion Coefficient $\tilde{d}$')

ax1.plot(d_tilda,AFC_nu_0(d_tilda,d0,F)*100)
ax1.set_title(r'$\eta_{AFC} = \tilde{d}^2 e^{\tilde{-d}} e^{\frac{-7}{F}} e^{{-d_0}}$')
ax1.set_xlabel(r'Absorbtion Coefficient $\tilde{d}$')
ax1.set_ylabel('Efficiency (%)')
ax1.grid('true')

ax2.plot(d_tilda,AFC_nu_1(d_tilda,F)*100)
ax2.set_title(r'$\eta_{AFC} = (1 - e^{\tilde{-d}})^2 e^{\frac{-7}{F^2}}$')
ax2.set_xlabel(r'Absorbtion Coefficient $\tilde{d}$')
ax2.set_ylabel('Efficiency (%)')
ax2.grid('true')

ax3.plot(d_tilda,AFC_nu_4(d_tilda,d0,F)*100)
ax3.set_title(r'$\eta_{AFC} = \tilde{d}^2 e^{\tilde{-d}} sinc(\frac{\pi}{F})^2 e^{{-d_0}}$')
ax3.set_xlabel(r'Absorbtion Coefficient $\tilde{d}$')
ax3.set_ylabel('Efficiency (%)')
ax3.grid('true')
```
